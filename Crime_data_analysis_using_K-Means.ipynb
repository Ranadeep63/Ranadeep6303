### Statistical Insights

```python
# Hypothesis Testing: ANOVA to check if there's a significant difference in Crime Rates across different areas

areas = merged_data['Area_Name'].unique()

```

## Detemine Optional Clusters

```python
# Selecting the relevant features for clustering

features = merged_data[['Victims_Above_50_Yrs_x', 'Victims_Between_10to14_Yrs', 'Victims_Between_14to18_Yrs', 
                        'Victims_Between_18to30_Yrs', 'Victims_Between_30to50_Yrs', 'Victims_of_Rape_Total', 
                        'Victims_Upto_10_Yrs_x', 'Total_Victims_By_Age_x', 'Crime_Rate_Rape', 
                        'Victims_Above_50_Yrs_y', 'Victims_Upto_10_15_Yrs', 'Victims_Upto_10_Yrs_y', 
                        'Victims_Upto_15_18_Yrs', 'Victims_Upto_18_30_Yrs', 'Victims_Upto_30_50_Yrs', 
                        'Victims_Total', 'Total_Victims_By_Age_y', 'Crime_Rate_Murder']]

```

### Elbow Method

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean of each column

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)

# Elbow Method for optimal k

distortions = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_imputed)
    distortions.append(kmeans.inertia_)

# Plotting the Elbow Method

plt.plot(range(1, 11), distortions, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

### silhouette Scores

```python
# Impute missing values with the mean of each column

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)

# Silhouette Scores for cluster quality validation

sil_scores = []
for k in range(2, 11):  # Silhouette score is undefined for k=1, so we start from k=2
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_imputed)
    sil_score = silhouette_score(features_imputed, kmeans.labels_)
    sil_scores.append(sil_score)

# Plotting Silhouette Scores

plt.plot(range(2, 11), sil_scores, marker='o')
plt.title('Silhouette Scores for Different k')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Find the optimal k based on the highest Silhouette Score

optimal_k = range(2, 11)[sil_scores.index(max(sil_scores))]
print(f"The optimal number of clusters based on Silhouette Score is: {optimal_k}")
```

## Hyperparameter Tunning & Cross-Validation

### Hyperparameter Tuning with Grid Search

```python
# Define the parameter grid

param_grid = {
    'n_clusters': [2, 3, 4, 5, 6],
    'init': ['k-means++', 'random'],  # Initialization methods
    'max_iter': [300, 500, 1000],  # Maximum iterations
}

# KMeans model
kmeans = KMeans(random_state=42)

# Grid search with cross-validation

grid_search = GridSearchCV(estimator=kmeans, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(features_imputed)

# Best parameters from grid search

print(f"Best Parameters: {grid_search.best_params_}")

# Best model from grid search

best_kmeans = grid_search.best_estimator_

# Fit the best model to the data

best_kmeans.fit(features_imputed)

# Add the predicted clusters to the data

merged_data['Best_Cluster'] = best_kmeans.labels_
```

### Cross-Validation


```python
# Define the KMeans model with the best parameters from GridSearch

kmeans = KMeans(n_clusters=grid_search.best_params_['n_clusters'],
                init=grid_search.best_params_['init'],
                max_iter=grid_search.best_params_['max_iter'],
                random_state=42)

# Initialize cross-validation

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# List to store silhouette scores

sil_scores = []

# Perform cross-validation and calculate silhouette score for each fold

for train_index, test_index in cv.split(features_imputed, np.zeros(len(features_imputed))):  # For unsupervised, use dummy labels
    X_train, X_test = features_imputed[train_index], features_imputed[test_index]
    
    # Fit KMeans on the training set
    
    kmeans.fit(X_train)
    
    # Predict the clusters for the test set
    
    cluster_labels = kmeans.predict(X_test)
    
    # Calculate silhouette score for the test set
    
    sil_score = silhouette_score(X_test, cluster_labels)
    sil_scores.append(sil_score)

# Print the cross-validation silhouette scores

print(f"Cross-validation silhouette scores: {sil_scores}")
print(f"Mean silhouette score: {np.mean(sil_scores)}")
```

### Dunn Index

```python
# Compute intra-cluster distances (compactness)

def intra_cluster_distance(data, labels):
    intra_distances = []
    for cluster in np.unique(labels):
        cluster_data = data[labels == cluster]
        centroid = cluster_data.mean(axis=0)
        distance = np.mean(np.linalg.norm(cluster_data - centroid, axis=1))
        intra_distances.append(distance)
    return np.mean(intra_distances)

# Compute inter-cluster distances (separation)

def inter_cluster_distance(data, labels):
    centroids = np.array([data[labels == cluster].mean(axis=0) for cluster in np.unique(labels)])
    distances = np.linalg.norm(centroids[:, np.newaxis] - centroids, axis=2)
    np.fill_diagonal(distances, np.inf)  # Ignore self-to-self distances
    return np.min(distances)

intra_distance = intra_cluster_distance(features_imputed_df.drop('cluster', axis=1), features_imputed_df['cluster'])
inter_distance = inter_cluster_distance(features_imputed_df.drop('cluster', axis=1), features_imputed_df['cluster'])

dunn_index = inter_distance / intra_distance
print(f'Dunn Index: {dunn_index}')
```
